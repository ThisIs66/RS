### Removing Punctuations and Stopwords

```python
from nltk.tokenize import word_tokenize

df['title_list'] = df['title'].str.lower()
df['actors_list'] = df['actors_list'].str.lower()
df['genre'] = df['genre'].str.lower()

df['title_list'] = df['title_list'].apply(word_tokenize)
df['actors_list'] = df['actors_list'].apply(word_tokenize)
df['genre'] = df['genre'].apply(word_tokenize)
```

```python
from nltk.corpus import stopwords
from string import punctuation

list_stopwords = set(stopwords.words('english') + list(punctuation))
df['title_list'] = df['title_list'].apply(lambda x: [word for word in x if word not in list_stopwords])
df['actors_list'] = df['actors_list'].apply(lambda x: [word for word in x if word not in list_stopwords])
df['genre'] = df['genre'].apply(lambda x: [word for word in x if word not in list_stopwords])
```

```python
import string

df['actors_list'] = df['actors_list'].apply(lambda x : [word.translate(str.maketrans('', '', string.punctuation)) for word in x])
df['actors_list'] = df['actors_list'].apply(lambda x : [word for word in x if len(word) > 0])
```

```python
df['title_list'] = df['title_list'].apply(lambda x : list(set(x)))
df['actors_list'] = df['actors_list'].apply(lambda x : list(set(x)))
df['genre'] = df['genre'].apply(lambda x : list(set(x)))
```
